```python
data.isnull().sum()
```

```python
data_c = pd.read_excel('Country-code.xlsx')
data_c.head()
```

```python
data_c.shape
```

```python
final = pd.merge(data,data_c,on='Country Code',how='left')
```

```python
final.sample(5)
```

```python
country_names = final.Country.value_counts().index
```

```python
country_num = final.Country.value_counts().values
```

```python
# Top 3 country using zomato
plt.figure(figsize=(14,7))
plt.pie(country_num[:3],labels=country_names[:3]
       ,autopct="%1.1f%%",shadow=True,colors=["brown","grey","w"])
plt.legend(loc='upper right')
plt.title('Countries using Zomato')
plt.show()
```

We see that number of transactions in India is significantly higher than that of in UK, US and other countries.

```python
num_ratings = final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Count'})
```

```python
num_ratings
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Rating text",y="Count",data=num_ratings,palette=['grey','red','orange','yellow','lightgreen','forestgreen'])
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Aggregate rating",y="Count",data=num_ratings)
plt.show()
```

- Most of the people don't rate about the service. 
- If we exclude Not rated category, The graph almost follow Normal Distribution.

```python
cities_name=final['City'].value_counts().index
```
