# Feature Engineering

We see that Date_of_Journey has dtype of object so we convert it into Date_time format.


```python
final['Date_of_Journey'].str.split('/')[:5]
```

```python
final['Date_of_Journey'].str.split('/').str[0]
```

```python
final['Date'] = final['Date_of_Journey'].str.split('/').str[0]
final['Month'] = final['Date_of_Journey'].str.split('/').str[1]
final['Year'] = final['Date_of_Journey'].str.split('/').str[2]
```

```python
final.head()
```

```python
final['Date'] = final['Date'].astype(int)
final['Month'] = final['Month'].astype(int)
final['Year'] = final['Year'].astype(int)
```

```python
final.info()
```

```python
final['Date_of_Journey'] = pd.to_datetime(final['Date_of_Journey'],format='%d/%m/%Y')
```

```python
final.info()
```

```python
final['Arrival_Time'] = final['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

```python
final.head()
```

```python
final['Arrival_Hr'] = final['Arrival_Time'].str.split(':').str[0]
```

```python
final['Arrival_Min'] = final['Arrival_Time'].str.split(':').str[1]
```

```python
final['Arrival_Hr'] = final['Arrival_Hr'].astype(int)
final['Arrival_Min'] = final['Arrival_Min'].astype(int)
```

```python
final.head()
```

```python
final.drop('Arrival_Time',axis =1 ,inplace= True)
```

```python
final['Dep_hr'] = final['Dep_Time'].str.split(':').str[0]
final['Dep_min'] = final['Dep_Time'].str.split(':').str[1]
```

```python
final['Dep_hr'] = final['Dep_hr'].astype(int)
final['Dep_min'] = final['Dep_min'].astype(int)
```

```python
final.drop('Dep_Time',axis = 1,inplace = True)
```

```python
final.head()
```

```python
# Deleting Route and focusing on Total_Stops.
final.drop('Route',axis=1,inplace=True)
```

```python
final.head()
```

```python
final.Total_Stops.value_counts()
```

```python
final['Total_Stops'] = final['Total_Stops'].map({'1 stop':1,'non-stop':0,'2 stops':2,'3 stops':3,'4 stops':4,'nan':0})
```

```python
final.info()
```

## Duration into minutes.

```python
final['dh'] = final['Duration'].str.split('h').str[0]
```

```python
final['dm'] = final['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
final['dh'].value_counts()
```

```python
# Removing 5m which is not possible.
final[final['dh']=='5m']
```

```python
final.drop(6474,axis=0,inplace = True)
final.drop(2660,axis=0,inplace = True)
```

```python
final['dm'].value_counts()
```

```python
# Converting into int.

final['dh'] = final['dh'].astype(int)
```

```python
final['dm'].fillna(0,inplace=True)
```

```python
final['dm'] = final['dm'].astype(int)
```

```python
final.head()
```

```python
final['Duration_In_Mins'] = final['dh']*60+final['dm'] 
```

```python
final.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
final['Airline'] = encoder.fit_transform(final['Airline'])
final['Source'] = encoder.fit_transform(final['Source'])
final['Destination'] = encoder.fit_transform(final['Destination'])
final['Additional_Info'] = encoder.fit_transform(final['Additional_Info'])
```

```python
final.head()
```

```python
final.drop('Duration',axis=1,inplace=True)
```
